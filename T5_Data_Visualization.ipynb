```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# **Read Data**

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

# **Basic Exploration**


1. Plot a histogram to visualize the distribution of math scores.

```python
plt.hist(df["Math_Score"], bins =10)
plt.xlabel("Math score")
plt.ylabel("Havanakanutyuny")
plt.title("Distribution of Math Scores")
plt.show()
```

2. Plot histograms for all three scores (math, reading, writing) side by side using subplots.


```python
fig, axes = plt.subplots(1,3, figsize = (15,5))

axes[0].hist(df["Math_Score"], bins = 10, edgecolor = "black")
axes[0].set_title("Math Scores")
axes[0].set_xlabel("Score")
axes[0].set_ylabel("Frequency")


axes[1].hist(df["Reading_Score"], bins = 10, edgecolor = "black")
axes[1].set_title("Reading_Score")
axes[1].set_xlabel("Score")
axes[1].set_ylabel("Frequency")


axes[2].hist(df["Writing_Score"], bins = 10, edgecolor = "black")
axes[2].set_title("Writing_Score")
axes[2].set_xlabel("Score")
axes[2].set_ylabel("Frequency")

```

3. Create a count plot to show the number of students in each gender group.


```python
sns.countplot(x= "Gender", data=df,palette="pastel",edgecolor ="black")
plt.title("Number of Students by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

4. Create a count plot for test preparation course grouped by gender using hue.


```python
sns.countplot(x="Test_Preparation_Course", hue="Gender", data=df,
              palette="pastel", edgecolor="black")
plt.title("Test Preparation Course by Gender")
plt.xlabel("Test Preparation Course")
plt.ylabel("Count")
plt.show()
```

# **Comparison by Category**

5. Use a bar plot to show the average writing score for each parental level of education.

```python
sns.barplot(x="Parental_Level_of_Education",
            y="Writing_Score",
            data=df,
            estimator="mean",

            edgecolor="black")

plt.title("Average Writing Score by Parental Level of Education")
plt.xlabel("Parental Level of Education")
plt.ylabel("Average Writing Score")
plt.xticks(rotation=45)
plt.show()
```

6. Create a box plot of reading scores grouped by gender.

```python
sns.boxplot(x="Gender", y="Reading_Score", data=df, palette="pastel")
plt.title("Reading Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Reading Score")

plt.show()
```

7. Create a violin plot of math scores grouped by lunch type.

```python
sns.violinplot(x="Lunch", y="Math_Score",data= df)
plt.title("Lunch Type by Math Score")
plt.xlabel("Lunch Type")
plt.ylabel("Math Score")
plt.show()
```

8. Plot a strip plot to show the distribution of math scores by race/ethnicity.

```python
sns.stripplot(x="Race/Ethnicity", y="Math_Score", data=df,
              palette="pastel", jitter=True, edgecolor="gray")

plt.title("Distribution of Math Scores by Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Math Score")

plt.show()
```

# **Correlation & Relationships**

 9. Plot a scatter plot of math score vs writing score, colored by gender.

```python
sns.scatterplot(x="Math_Score", y="Writing_Score", hue="Gender", data=df,
                palette="pastel", s=80, edgecolor="black")

plt.title("Math Score vs Writing Score by Gender")
plt.xlabel("Math Score")
plt.ylabel("Writing Score")

plt.show()
```

10. Add a regression line to that scatter plot using sns.lmplot.

```python
sns.lmplot(x="Math_Score", y="Writing_Score", hue="Gender", data=df,
           palette="pastel")

plt.title("Math Score vs Writing Score with Regression Line by Gender")
plt.xlabel("Math Score")
plt.ylabel("Writing Score")

plt.show()
```

11. Create a pair plot for math, reading, and writing scores, colored by gender.

```python
sns.pairplot(df, vars=["Math_Score", "Reading_Score", "Writing_Score"],
             hue="Gender", palette="pastel", diag_kind="kde")

plt.suptitle("Pair Plot of Scores by Gender", y=1.02)
plt.show()
```

12. Create a heatmap of the correlation matrix for the scores.

```python
score_cols = ["Math_Score", "Reading_Score", "Writing_Score"]
corr_matrix = df[score_cols].corr()

sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

plt.title("Correlation Matrix of Scores")
plt.show()
```

13. Group by race/ethnicity, then plot a line plot showing average scores for each subject.


```python
grouped = df.groupby("Race/Ethnicity")[["Math_Score", "Reading_Score", "Writing_Score"]].mean()

plt.figure(figsize=(8,5))
for subject in grouped.columns:
    plt.plot(grouped.index, grouped[subject], marker='o', label=subject)

plt.title("Average Scores by Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Average Score")
plt.legend()
plt.grid(True)
plt.show()
```

14. Create a facet grid of histograms, one per race group, for math scores.


```python
g = sns.FacetGrid(df, col="Race/Ethnicity", col_wrap=3, height=4, sharex=True, sharey=True)
g.map(plt.hist, "Math_Score", bins=10, color="skyblue", edgecolor="black")

g.set_titles(col_template="{col_name}")
g.set_axis_labels("Math Score", "Count")
g.fig.suptitle("Distribution of Math Scores by Race/Ethnicity", y=1.05)

plt.show()
```

15. Create a bar plot with transparent bars and colored outlines (custom facecolor, edgecolor).


```python
avg_scores = df.groupby("Gender")["Math_Score"].mean()
plt.bar(avg_scores.index, avg_scores.values,
        facecolor="skyblue",
        edgecolor="darkblue",
        linewidth=2,
        alpha=0.5)

plt.title("Average Math Scores by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Math Score")
plt.show()
```

16. Sort the parental level of education by average reading score before plotting.


```python
avg_reading = df.groupby("Parental_Level_of_Education")["Reading_Score"].mean().sort_values()


order = avg_reading.index.tolist()

sns.barplot(x="Parental_Level_of_Education", y="Reading_Score", data=df,
            order=order, palette="pastel", edgecolor="black")

plt.title("Average Reading Score by Parental Level of Education (Sorted)")
plt.xlabel("Parental Level of Education")
plt.ylabel("Average Reading Score")
plt.xticks(rotation=45)
plt.show()
```

17. Create a pie chart showing the distribution of lunch types.


```python
lunch_counts = df["Lunch"].value_counts()


plt.figure(figsize=(6,6))
plt.pie(lunch_counts, labels=lunch_counts.index, autopct="%1.1f%%",
        colors=["skyblue", "lightgreen"], startangle=90, explode=(0.05, 0.05),
        shadow=True)

plt.title("Distribution of Lunch Types")
plt.show()
```

18. Create a stacked bar chart showing counts of students by gender and test prep course.


```python
counts = df.groupby(["Gender", "Test_Preparation_Course"]).size().unstack()

counts.plot(kind="bar", stacked=True, color=["skyblue", "lightgreen"], edgecolor="black", figsize=(8,5))

plt.title("Counts of Students by Gender and Test Preparation Course")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.legend(title="Test Preparation Course")
plt.xticks(rotation=0)
plt.show()
```
